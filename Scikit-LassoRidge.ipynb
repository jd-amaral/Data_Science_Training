```python
import pandas as pd
```

```python
#archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data
auto_data = pd.read_csv('datasets/02/demos/data/imports-85.data', sep=r'\s*,\s*', engine='python')
auto_data
```

```python
import numpy as np

auto_data = auto_data.replace('?', np.nan)
auto_data.head()
```

```python
auto_data.describe(include='all')
```

```python
auto_data['price'].describe() #is object and needs to be converted to numeric
```

```python
auto_data['price'] = pd.to_numeric(auto_data['price'], errors='coerce')
```

```python
auto_data['price'].describe()
```

```python
auto_data = auto_data.drop('normalized-losses', axis=1)
auto_data.head()
```

```python
auto_data['horsepower'] = pd.to_numeric(auto_data['horsepower'], errors='coerce')
auto_data['horsepower'].describe()
```

```python
auto_data['num-of-cylinders'].describe()
cylinders_dict = {
    'two': 2,
    'three': 3,
    'four': 4,
    'five': 5,
    'six': 6,
    'eight': 8,
    'twelve': 12
}
auto_data['num-of-cylinders'].replace(cylinders_dict, inplace=True)
auto_data.head()
```

```python
auto_data = pd.get_dummies(auto_data, columns=['make', 'fuel-type','aspiration','num-of-doors', 'body-style', 'drive-wheels', 'engine-location', 'engine-type', 'fuel-system'])
auto_data.head()
```

```python
auto_data = auto_data.dropna()
auto_data
```

```python
auto_data[auto_data.isnull().any(axis=1)]
```

```python
from sklearn.model_selection import train_test_split

X = auto_data.drop('price', axis=1)

#Taking the labels (price)
Y = auto_data['price']

#Splitting into 80% training and 20% testing
X_train, x_test, Y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression

linear_model = LinearRegression()
linear_model.fit(X_train, Y_train)
```

```python
linear_model.score(X_train, Y_train) #R^2 value prediction
```

```python
linear_model.coef_
```

```python
predictors = X_train.columns #which features are being used
coef = pd.Series(linear_model.coef_, predictors).sort_values() #correlacion coefficients of the features
print(coef)
```

```python
y_predict = linear_model.predict(x_test)
```

```python
%pylab inline
pylab.rcParams['figure.figsize'] = (15, 6)

plt.plot(y_predict, label='Predicted')
plt.plot(y_test.values, label='Actual')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
r_square = linear_model.score(x_test, y_test)
r_square
```

```python
from sklearn.metrics import mean_squared_error

linear_model_mse = mean_squared_error(y_predict, y_test)
linear_model_mse
```

```python
import math 

math.sqrt(linear_model_mse)
```

```python
from sklearn.linear_model import Lasso
from sklearn.pipeline import Pipeline
import sklearn.preprocessing 

Xtrain_ = sklearn.preprocessing.normalize(X_train, norm="l1")
lasso = Lasso(alpha= 2, random_state=0)
lasso.fit(Xtrain_, Y_train)

pipe = Pipeline(steps= [
    ('scaler', sklearn.preprocessing.Normalizer()), 
    ('lasso', Lasso(alpha = 2, random_state=0))
])

pipe.fit(Xtrain_, Y_train)


```

```python
pipe.score(Xtrain_, Y_train) 
```

```python
lasso_coef = pd.Series(pipe.named_steps['lasso'].coef_,predictors).sort_values()
print(lasso_coef)
```

```python
y_predict = pipe.predict(x_test)
```
